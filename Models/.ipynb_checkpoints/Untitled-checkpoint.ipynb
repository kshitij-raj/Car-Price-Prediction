```python
#import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
#read the dataset
df = pd.read_csv("F:\CarPricePrediction\Dataset\car data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#convert Selling Price and Present price of the car from Lakhs to thousand

df['Selling_Price'] = df['Selling_Price'].apply(lambda x: x*100000)
df['Present_Price'] = df['Present_Price'].apply(lambda x: x*100000)
```

```python
#Drop columns which are not required.
df = df.drop('Car_Name',axis=1)
```

```python
# Derived Metrics

##Metrics to define age of the car
current_year = 2021
df['Car_Age'] =  df['Year'].apply(lambda x: current_year - x)

#Drop Year column
df = df.drop('Year',axis=1)
```

```python
#Create dummpy variabe
df = pd.get_dummies(df,drop_first=True)
df.head()
```

```python
#Correlation Plot
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
#Create dependent and independent features.
X = df.loc[:,df.columns != 'Selling_Price']
y = df.loc[:,df.columns == 'Selling_Price']
```

```python
#Train-Test Spit
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=40,shuffle=True)

print("X_train shape - ",X_train.shape)
print("X_test  shape - ",X_test.shape)
```

```python
#Model Training
randomforest  = RandomForestRegressor()

#Hyperparameter tuning
rfparams = {'n_estimators': [10,25,50,100],
            'max_depth': [3,5],
            'min_samples_leaf': [2,5,10,50],
            'max_features': ['auto', 'sqrt']}
model_randomforest = RandomizedSearchCV(estimator=randomforest,param_distributions=rfparams,n_iter=10,cv=5,verbose=1,n_jobs=-1,scoring="neg_mean_squared_error",random_state=42)
```

```python
model_randomforest.fit(X_train,y_train)
```

```python
#best parameter
model_randomforest.best_estimator_
```

```python
#RandomForest model with best hyper parameter value.
model_randomforestbest = model_randomforest.best_estimator_

#predict the target variable.
y_test_pred =model_randomforestbest.predict(X_test)
```

```python
sns.distplot(y_test_pred)
plt.show()
```

```python
import pickle

#Open the file where you want to store the data.
file = open('randoforest_regression_model.pkl','wb')

#dump information to that file.
pickle.dump(model_randomforestbest,file)
```
